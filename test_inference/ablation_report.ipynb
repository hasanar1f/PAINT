```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame with the desired data
data = {
    "VQAv2": [77.96, 79.72, 79.16, 79.2, 79.4, 77.76, 77.60, 77.56, 73.95],
    "TextVQA": [50.24, 53, 56, 58.64, 57.76, 57.64, 59.32, 59.44, 52.24],
    'ScienceQA': [68, 68.4, 69.6, 69.6, 70, 68.4, 68.4, 68.4, 65.6]
}

# Define the custom index
custom_index = [1, .875, .75, .625, .50, .375, .25, .125, 0]

# Create the DataFrame
df = pd.DataFrame(data, index=custom_index)

# Plot the data with size 8x8
fig, ax = plt.subplots(figsize=(6, 5))
df.plot(ax=ax)

# Add labels and title with text size 16
ax.set_xlabel(r'Value of $\alpha$', fontsize=16)
ax.set_ylabel('Accuracy', fontsize=16)
ax.legend(fontsize=16)

# Set the y-axis limit to start from 35
ax.set_ylim(35, 85)

# Draw a vertical dotted line at x = 0.5 and label it as "default"
ax.axvline(x=0.5, color='gray', linestyle='--')
# set label for the line
ax.text(0.52, 40, r'$\alpha=0.5$ as default', rotation=0, verticalalignment='bottom', fontsize=16)
# ax.text(0.5, 80, 'default', rotation=0, verticalalignment='bottom', fontsize=16)

# Adjust the layout for better fit
plt.tight_layout()

# Show the plot
# save as pdf
plt.savefig('alpha.pdf')
plt.show()
```
